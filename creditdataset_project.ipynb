```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.github.com/Premalatha-success/Yhills_July12_Analytics/main/credit.csv")
```

```python
data['checking_balance'].replace('< 0 DM',0,inplace=True)
data['checking_balance'].replace('1 - 200 DM',100,inplace=True)
data['checking_balance'].replace('> 200 DM',300,inplace=True)
data['checking_balance'].replace('unknown',np.nan,inplace=True)
data['checking_balance']=data['checking_balance'].astype(float)
```

```python
median1=data['checking_balance'].median()
data['checking_balance'].replace(np.nan,median1,inplace=True)
```

```python
data.sample(10)
```

```python
data['savings_balance'].replace('< 100 DM',100,inplace=True)
data['savings_balance'].replace('100 - 500 DM',250,inplace=True)
data['savings_balance'].replace('500 - 1000 DM',750,inplace=True)
data['savings_balance'].replace('> 1000 DM',1500,inplace=True)
data['savings_balance'].replace('unknown',np.nan,inplace=True)
data['savings_balance']=data['savings_balance'].astype(float)
```

```python
median2=data['savings_balance'].median()
data['savings_balance'].replace(np.nan,median2,inplace=True)
```

```python
data['employment_duration'].replace('< 1 year',1,inplace=True)
data['employment_duration'].replace('1 - 4 years',2.5,inplace=True)
data['employment_duration'].replace('4 - 7 years',5.5,inplace=True)
data['employment_duration'].replace('> 7 years',8,inplace=True)
data['employment_duration'].replace('unemployed',0,inplace=True)
data['employment_duration']=data['employment_duration'].astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale=StandardScaler()
std_scale
```

```python
data['checking_balance']=std_scale.fit_transform(data[['checking_balance']])
#data['credit_history']=std_scale.fit_transform(data[['credit_history']])
#data['purpose']=std_scale.fit_transform(data[['purpose']])
data['amount']=std_scale.fit_transform(data[['amount']])
data['savings_balance']=std_scale.fit_transform(data[['savings_balance']])
data['employment_duration']=std_scale.fit_transform(data[['employment_duration']])
data['percent_of_income']=std_scale.fit_transform(data[['percent_of_income']])
data['years_at_residence']=std_scale.fit_transform(data[['years_at_residence']])
data['age']=std_scale.fit_transform(data[['age']])
#data['other_credit']=std_scale.fit_transform(data[['other_credit']])
#data['housing']=std_scale.fit_transform(data[['housing']])
data['existing_loans_count']=std_scale.fit_transform(data[['existing_loans_count']])
#data['job']=std_scale.fit_transform(data[['job']])
data['dependents']=std_scale.fit_transform(data[['dependents']])
#data['phone']=std_scale.fit_transform(data[['phone']])
#data['default']=std_scale.fit_transform(data[['default']])
```

```python
dummies=pd.get_dummies(data[['credit_history','purpose','other_credit','housing','job','phone','default']],
                      columns=['credit_history','purpose','other_credit','housing','job','phone','default'])
dummies.head(10)
```

```python
columns=['credit_history','purpose','other_credit','housing','job','phone','default']
data=pd.concat([data,dummies],axis=1)
data.drop(columns,axis=1,inplace=True)
```

```python
data.head(10)
```

```python
data.sample(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=data.drop(['months_loan_duration'],axis=1)
y=data[['months_loan_duration']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
reg_model=LogisticRegression()
reg_model.fit(x_train,y_train)
predictions=reg_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```

```python
reg_model.score(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC(kernel='sigmoid',C=2.0,gamma='auto')
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
clf.score(x_test,y_test)
```
